```python
# dictionary problem 1
d = {
    "a":[5,71,60],
    "b":[6,66,76],
    "c":[1,81,97],
    "d":[8,99,87],
    "e":[7,90,92]
}
name = input()
if(d[name][2]>75):
  print("strong perfomance")
elif(d[name][2]>=40):
  print("need improvement")
else:
  print("repeat class")

if(d[name][1]<75):
  print("warning-low attendance")

if(d[name][0]<2):
  print("warning-study more")
# output:
#  c
#  strong perfomance
#  warning-study more

```

```python
# dictionary problem 2
# student starter code
transactions = [
    {"transaction_id": "TX101", "amount": 6200, "method": "Credit Card"},
    {"transaction_id": "TX102", "amount": 5500, "method": "Gift Card"},
    {"transaction_id": "TX103", "amount": 1200, "method": "Gift Card"},
    {"transaction_id": "TX104", "amount": 8000, "method": "Gift Card"},
    {"transaction_id": "TX105", "amount": 450, "method": "PayPal"}
]

flagged_transactions = []

# Write your code below to find anomalous patterns|


for i in range(len(transactions)):
  if(transactions[i]["amount"]>5000 and transactions[i]["method"]=="Gift Card"):
    flagged_transactions.append(transactions[i]["transaction_id"])

for i in flagged_transactions:
  print(i)
# TX102
# TX104

```
