```python
import pandas as pd
```

```python
df=pd.read_csv("session4_dummy.csv")
```

```python
print(df.head())
print(df.columns)
print(df.shape)
print(df.describe())
```

```python
math_scores=df["math"].to_numpy()
print(math_scores)
print(type(math_scores))
print(math_scores.mean())
```

```python
#calculate the average score across the other two subjects
#filtering
#top students average more than or equal to 85
df["average"]=df[["math","science","english"]].mean(axis=1)
print(df[["name","math","science","english","average"]])

top_students=df[df["average"]>=85]
print(top_students[["name","average"]])
```

```python
#Making filters
#low attendance,if attendance percentage is less than 75 then
low_attendance=df[df["attendance_percent"]<75]
print(low_attendance[["name","attendance_percent"]])
```

```python
#if students are from Mathura....display their name,city and average
mathura_students=df[df["city"]=="Mathura"]
print(mathura_students[["name","city","average"]])
```

1.Read session4_dummy_students.csv.

2.Create an average column using math, science, and english.

3.Create a result column: Pass if average >= 40, otherwise Fail.

4.Print class average.

5.Print highest average and lowest average.

6.Print top student or students.

7.Print students with attendance below 75.

8.Print students who submitted fewer than 6 assignments.

9.Save final output as session4_student_report.csv.

```python
#1.
dff=pd.read_csv("session4_dummy.csv")
```

```python
#2.
dff['average'] = df[['math', 'science', 'english']].mean(axis=1)
dff
```

```python
#3.
dff['result'] = dff['average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
```

```python
#4.
class_avg = dff['average'].mean()
print("Class Average:", class_avg)
```

```python
#5.
high = dff['average'].max()
low = dff['average'].min()
print("Highest Average:", high)
print("Lowest Average:", low)
```

```python
#6.
top= dff[dff['average'] == high]
print(top['name'])
```

```python
#7.
low_attendance = dff[dff['attendance_percent'] < 75]
print(low_attendance['name'])
```

```python
#8.
names=dff[dff['attendance_percent'] < 75]['name']
print(names)
```

```python
#9.
dff.to_csv('session4_newReport.csv', index=False)
```
